In dieser Aufgabe sollen verschiedene Summenfunktionen berechnet werden. Jeweils einmal mit und einmal ohne Schleife. 

#### Mit Schleife

* range(1001) erzeugt eine Sequenz von Zahlen von 0 bis 1000 (Die Zahl 1001 wird gewählt, weil in Python das ﻿range-Objekt bis zur angegebenen Zahl exklusiv läuft, was bedeutet, dass ﻿range(1001) tatsächlich bis 1000 läuft)
* for k in range(1001) iteriert über jede Zahl k in dieser Sequenz
* k**2 berechnet das Quadrat von k
* die Summenfunktion sum() addiert alle quadrierten Werte für k von 0 bis 1000

#### Ohne Schleife
Die Berechnung ohne Schleife erfolgt mittels Numpy

* np.arange(1001) ergibt ein Array von 0 bis 1000
* np.arange(1001) ** 2 quadriert jedes Element des Arrays elementweise
* np.sum() summiert alle quadrierten Werte im Array






```python
import numpy as np

result = sum(k**2 for k in range(1001))
print(f"Mit Schleife: {result}")

result = np.sum(np.arange(1001) ** 2)
print(f"Ohne Schleife: {result}")
```

#### Rundungsfehler:

Nur Maschinenzahlen können exakt im Computer dargestellt werden. Alle anderen Werte müssen zur nächstliegenden Maschinenzahl gerundet werden, wobei **Rundungsfehler** auftreten. Wenn man Fließkommazahlen als double darstellt, sind diese auf 16 Ziffern genau (*double precision*).

Rundungsfehler aufgrund der Maschinengenauigkeit treten also in der **16. Dezimalstelle** auf.

Die Rundungsfehler werden dementsprechend in Aufgabe c.sichtbar, da hier im Ergebnis die 16. Dezimalstelle erreicht wird.

#### Methode Mit und Ohne Vorschleife liefert unterschiedliche Ergebnisse, warum?

##### Mit Schleife

Die Methode mit einer Schleife, folgt einer **sequentiellen Logik**. D.h, der vorherig berechnete Wert dient als Ausgangswert für die nächste Berechnung. Demnach verstärkt sich der Rundungsfehler je länger die Schleife läuft.

##### Ohne Schleife

Die zweite Methode mit np.sum ohne Schleife ist genauer, weil np.sum eine **vektorisierte Berechnung** durchführt und somit potenzielle Rundungsfehler aufgrund von vorherigen Rundungsfehlern minimiert.

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 5 * np.exp(x / 3)

def g(x):
    return 0.1 * x**4


def plotting(function):

    x = np.linspace(0.01, 20, 500)
    y = function(x)

    fig, axs = plt.subplots(2, 2, figsize=(12, 8))

    # Normal
    axs[0, 0].plot(x, y)
    axs[0, 0].set_title('Normal Scale')
    axs[0, 0].set_xlabel('x')
    axs[0, 0].set_ylabel(f'{function}(x)')
    axs[0, 0].grid(True)

    # Log X
    axs[0, 1].semilogx(x, y)
    axs[0, 1].set_title('Logarithmic X Scale')
    axs[0, 1].set_xlabel('x')
    axs[0, 1].set_ylabel(f'{function}(x)')
    axs[0, 1].grid(True)

    # Log Y
    axs[1, 0].semilogy(x, y)
    axs[1, 0].set_title('Logarithmic Y Scale')
    axs[1, 0].set_xlabel('x')
    axs[1, 0].set_ylabel(f'{function}(x)')
    axs[1, 0].grid(True)

    # Log X and Y
    axs[1, 1].loglog(x, y)
    axs[1, 1].set_title('Log-Log Scale')
    axs[1, 1].set_xlabel('x')
    axs[1, 1].set_ylabel(f'{function}(x)')
    axs[1, 1].grid(True)

    plt.tight_layout()
    plt.show()

plotting(f)
```


### b)

```python
plotting(g)
```

## Aufgabe 3
### a)

```python
def log_next(x, r):
    return r * x * (1 - x)

x = 0.1
r = 1.5
n = 20

for i in range(n):
    x = log_next(x, r)
    print(f"x{i+1}: {x}")
```

### b)

```python
def logsequence(x, r, n):

    sequence = [x]
    for i in range(n):
        x = log_next(x, r)
        sequence.append(x)
    return sequence

print(logsequence(0.1, 1.5, 20))
    
```

### c)

```python
def longterm_development(r,rand_min_x0,rand_max_x0):
    random_values = np.random.uniform(rand_min_x0,rand_max_x0 , 5)

    for rand_x0 in random_values:
        log_sequences = logsequence(rand_x0, r, 30)
        plt.plot(log_sequences, label=f"x0 = {rand_x0}")

    plt.grid(True)
    plt.xlabel('n')
    plt.ylabel('x')
    plt.legend()

longterm_development(2.5, 0, 1)

```
